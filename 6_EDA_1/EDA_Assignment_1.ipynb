### Objective #3.1: What cities are in the data?

First I checked the V2/countries and found 2 in the cities column. 

```python
# v2 countries
# API Address with parameters set
PATH_API = "https://u50g7n0cbj.execute-api.us-east-1.amazonaws.com/v2/countries?limit=200&page=1&offset=0&sort=asc&country_id=SA&order_by=country"

# Make a GET request
r = requests.get(PATH_API)

# Put the data into a pandas DataFrame
pd.DataFrame(r.json()["results"])
```

So then I checked the V2/cities and found Dhahran and NA in the city column. 

```python
# v2/cities
# API Address with parameters set
PATH_API2 = "https://u50g7n0cbj.execute-api.us-east-1.amazonaws.com/v2/cities?limit=100&page=1&offset=0&sort=asc&country_id=SA&order_by=city"

# Make a GET request
r2 = requests.get(PATH_API2)

# Put the data into a pandas DataFrame
pd.DataFrame(r2.json()["results"])
```

So then I went to the V2/locations and found the cities in the name column as you can see below: 

```python
# v2 locations
# API Address with parameters set
PATH_API3 = "https://u50g7n0cbj.execute-api.us-east-1.amazonaws.com/v2/locations?limit=100&page=1&offset=0&sort=desc&radius=1000&country_id=SA&order_by=lastUpdated&dumpRaw=false"

# Make a GET request
r3 = requests.get(PATH_API3)

# Put the data into a pandas DataFrame
cities = pd.DataFrame(r3.json()["results"])
```
