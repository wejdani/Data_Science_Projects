### Method 1: pandas.DataFrame.apply()

#### Description:
This method allows the user to pass in a function that will be applied on every single value of the dataframe. 


#### Parameters: 
**func: *function***

Function to apply to each column or row.

**axis: *{0 or ‘index’, 1 or ‘columns’}, default 0***


Axis along which the function is applied:

* 0 or ‘index’: apply function to each column.

* 1 or ‘columns’: apply function to each row.

**raw: *bool, default False***


Determines if row or column is passed as a Series or ndarray object:

* False : passes each row or column as a Series to the function.

* True : the passed function will receive ndarray objects instead. If you are just applying a NumPy reduction function this will achieve much better performance.

**result_type: *{‘expand’, ‘reduce’, ‘broadcast’, None}, default None***


These only act when axis=1 (columns):

* ‘expand’ : list-like results will be turned into columns.

* ‘reduce’ : returns a Series if possible rather than expanding list-like results. This is the opposite of ‘expand’.

* ‘broadcast’ : results will be broadcast to the original shape of the DataFrame, the original index and columns will be retained.

The default behaviour (None) depends on the return value of the applied function: list-like results will be returned as a Series of those. However if the apply function returns a Series these are expanded to columns.

**args: *tuple***


Positional arguments to pass to func in addition to the array/series.


#### Returns: Series or DataFrame


Result of applying func along the given axis of the DataFrame.

#### Examples: 
##### Example #1: Sum all column values

```python
# first get the numeric columns 
num_zoo_df = zoo_df.select_dtypes(include=np.number)

# apply the numpy sum method to the dataframe using apply on axis=0 which means column 
print(num_zoo_df.apply(np.sum, axis=0)) 
```

##### Example #2: Sum all row values

```python
# first get the numeric columns 
num_zoo_df = zoo_df.select_dtypes(include=np.number)

# apply the numpy sum method to the dataframe using apply on axis=1 which means row 
print(num_zoo_df.apply(np.sum, axis=1))
```

### Method 2: pandas.DataFrame.nlargest()

#### Description: 

This function returns the *n* largest values of the columns specified. 
#### Parameters: 

**n: *int***
Number of rows to return.

**columns: *label or list of labels***
Column label(s) to order by.

**keep: *{‘first’, ‘last’, ‘all’}, default ‘first’***
Where there are duplicate values:

* *first* : prioritize the first occurrence(s)

* *last* : prioritize the last occurrence(s)

* *all*: **do not drop any duplicates, even it means**
selecting more than *n* items.

#### Return: DataFrame
The first *n* rows ordered by the given columns in descending order.

#### Examples: 
##### Example #1: get 3 records of the highest number of legs

```python
zoo_df.nlargest(3,'legs')
```

##### Example #2: get 3 records of the highest type

```python
zoo_df.nlargest(3,'type')
```

* Research and explain the difference between a pandas Series and a pandas DataFrame. List 2 methods that can only be used for a Series and not a DataFrame

#### Answer: 

##### pd.Series: 
A Pandas Series is a **one** dimensional data structure, it is compromised of a key value pair. It can be described as being similar to a python dictionary but with more flexibility to manipulate data. Because the key represents an index and the value is of type ndarray this allows for more functionality to be applied compared to dictionaries. 

<img  src="./series.PNG"> 


**Attributes of pd.Series:**
* Series is a labeled ndarray
* Index: series are indexed by default from 0 to *n* and indexes are customizable


##### pd.DataFrame: 
A Pandas Dataframe is a **two** dimensional data structure, it is compromised of rows and columns that resembles a table or a spreadsheet. It can be described as being a combination of 2 or more series. 

<img  src="./dataframe.PNG"> 

**Attributes of pd.DataFrames:**
* The data in a dataframe is stored in memory as a collection of series
* Index: dataframes are indexed by default from 0 to *n* and indexes are customizable
* 
##### Key takeaways: 

| pd.Series | pd.DataFrame |
| :- | :-|
| **1D** data structure | **2D** data structure |
| Key-Value pair | Spreadsheet |
| labeled ndarray | collection of series|

##### Methods that apply to series ONLY: 
* **map** maps a value of series according to input
* **to_numpy** returns an ndarray
* **to_list** returns a list of values
