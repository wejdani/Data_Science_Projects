```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

import plotly.express as px
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import random
import dash
import json
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
```

```python
AC = pd.read_csv('./data/AC/villagers.csv')
AC_IMG = pd.read_csv('./data/AC_IMG/villagers.csv')
```

```python
AC_cols = list(AC.columns)
AC_IMG_cols = list(AC_IMG.columns)
# get the unique names from both lists and then add 'Name' so we can merge both datasets
unique=[x for x in AC_IMG_cols if x not in AC_cols]
unique.append("Name")
# merge both datasets on Name
AC_IMG=AC_IMG[unique]
AC_Villagers=AC.merge(AC_IMG, on="Name")
```

```python
# Build App
app = JupyterDash(__name__)


# Create labebls for pie chart input
palette_labels = ['cloud', 'pastel', 'cottage', 'jenasais']
themes= ['ggplot2', 'none', 'plotly_dark','plotly_white', 'seaborn']

songs_list = []
songs_list.append(list(AC_Villagers["Favorite Song"].value_counts().index[0:20]))
songs_list.append(list(AC_Villagers["Favorite Song"].value_counts().index[21:40]))
songs_list.append(list(AC_Villagers["Favorite Song"].value_counts().index[41:60]))


        
top_list=["Top 3", "Top 5", "All"]          
app.layout = html.Div([
        dcc.Tabs([
        dcc.Tab(label='Pie Chart', children=[
            #Title and left and right icons
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'right','height':'40%'}),       
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'left','height':'40%'}),
            html.Center(html.H1("Pie Chart of Animal Crossing Villagers")),

                       # Area to hold Graph
                       dcc.Graph(id="graph", style={'padding-top':'50px'}),

                       # First Dynamic Input for Tab 1
                       html.Label([
                                   "Pie Chart",
                                    dcc.Dropdown(
                                        id='pie', 
                                        clearable=False,
                                        value= "Personality", 
                                        options=[{'label': "Personality", 'value': "Personality"},
                                                 {'label': "Gender", 'value': "Gender"},
                                                 {'label': "Style 1", 'value': "Style 1"},
                                                {'label': "Style 2", 'value': "Style 2"},
                                                {'label': "Hobby", 'value': "Hobby"},]
                                        )
                       ]),
                       
                       # Line Break
                       html.Br([]),

                       # Second Dynamic Input for Tab 1
                       html.Label([
                                   "Palette",
                                   dcc.RadioItems(
                                       id="palette",
                                       options=[{'label':x, "value":x} for x in palette_labels],
                                       value='pastel',
                                       labelStyle={'display': 'inline-block'}
                                    )  
                                                        ])
        ]),
        dcc.Tab(label='Scatter Plot', children=[
            #Title and left and right icons
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'right','height':'40%'}),              
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'left','height':'40%'}),
            html.Center(html.H1("Scatter Plots of Animal Crossing Villagers")),
                       # Area to hold Graph
                       dcc.Graph(id="graph2", style={'padding-top':'50px'}),

                       # First Dynamic Input for Tab 2
                       html.Label([
                                   "Variable",
                                    dcc.Dropdown(
                                        id='variable', 
                                        clearable=False,
                                        value= "Personality", 
                                        options=[{'label': "Personality", 'value': "Personality"},
                                                 {'label': "Gender", 'value': "Gender"},
                                                 {'label': "Style 1", 'value': "Style 1"},
                                                {'label': "Hobby", 'value': "Hobby"},]
                                        )
                       ]),
                       
                       # Line Break
                       html.Br([]),

                       # Second Dynamic Input for Tab 2
                       html.Label([
                                   "Themes",
                                   dcc.RadioItems(
                                       id="themes",
                                       options=[{'label':x, "value":x} for x in themes],
                                       value='ggplot2',
                                       labelStyle={'display': 'inline-block'}
                                    )  
                                                        ])
        ]),
                    dcc.Tab(label='Bar Plot', children=[
                        #Title and left and right icons
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'right','height':'40%'}),              
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'left','height':'40%'}),
            html.Center(html.H1("Bar Plot of Villagers Favorite Songs")),

                       # Area to hold Graph
                       dcc.Graph(id="graph3", style={'padding-top':'50px'}),
                        

                       # First Dynamic Input for Tab 3
                       html.Label([
                                   "Bar Chart",
                                    dcc.Dropdown(
                                        id='bar', 
                                        value= "Sample 1", 
                                        options=[{'label': "Sample 1", 'value': "Sample 1"},
                                                 {'label': "Sample 2", 'value': "Sample 2"},
                                                 {'label': "Sample 3", 'value': "Sample 3"}
                                                ]
                                        )
                       ]),
                       
                       # Line Break
                       html.Br([]),                  
                       # Second Dynamic Input for Tab 3
                       html.Label([
                                   "Top",
                                   dcc.RadioItems(
                                       id="top",
                                       options=[{'label':x, "value":x} for x in top_list],
                                       value='All',
                                       labelStyle={'display': 'inline-block'}
                                    )  
                                                        ])
        
                        
                        
                        
                    ]),
                    dcc.Tab(label='Villagers', children=[
                        #Title and left and right icons
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'right','height':'40%'}),              
            html.Img(src=AC_Villagers.loc[random.randint(1,392),"Icon Image"],style={'float': 'left','height':'40%'}),
            html.Center(html.H1("Get a random Villager Profile")),
                        
                        html.Center(html.Button('Click here to get a random Villager!', id='show-secret'), style={"padding":"15px"}),
                    html.Div(id='body-div')
            ]),


    ])


                       
])


# Note: if you have more than one input or output make sure to have them in a list
@app.callback(
    [Output('graph', 'figure'),
    Output('graph2', 'figure')],
    [Input("pie", "value"),
     Input("palette", "value"),
    Input("variable", "value"),
    Input("themes", "value")]
    
)
def update_figure(val_pie, val_palette, val_variable, val_themes):

    # Palettes
    cloud = ["#F08D7E" , "#EFA18A" ," #E2BAB1" , "#DDA6B9" , '#ACAEC5',"#9b8689","#ec9295","#fadddb"]
    cottage=["#596854" , "#7F803E" , "#CC9A52" , '#AD794B' , "#FCE4B4","#630d16","#2e191e",'#37432b']
    pastel=['#D8A7B1','#8FDDE7','#B6E2D3', '#FAE8E0', '#EF7C8E', '#FBE5C8', '#A49393',"#f56b5c"]
    jenasais=['#FF68A8' , '#64CFF7' ,'#F7E752' , '#CA7CD8' , '#3968CB','#c03836','#7bca76',"#dd5d42"]

    # set the palette from the input val_palette
    color_dict={'cloud':cloud, 'cottage':cottage, 'pastel':pastel, 'jenasais':jenasais}
    colors=[]
    for k,v in color_dict.items():
        if(val_palette==k):
            colors=v
        
        
    if  (val_pie == "Personality"):
        elements = AC_Villagers["Personality"].value_counts()     # slice data
        elements_labels = elements.index.tolist()                 # get the names for the labels from the index
        elements_values = elements.values                         # to set values from the dataframe
        pie= go.Figure(go.Pie(                                    # assign pie as our first output to return
        name = "", 
        values = elements_values,
        labels = elements_labels,
        marker=dict(colors=colors, line=dict(color='#000000', width=2)), # assign the color palette from colors list
        hovertemplate = "%{label}: <br>Amount: %{percent} </br>"
        ))
        
    
    elif  (val_pie == "Hobby"):
        elements = AC_Villagers["Hobby"].value_counts()           # slice data
        elements_labels = elements.index.tolist()                 # get the names for the labels from the index
        elements_values = elements.values                         # to set values from the dataframe
        pie= go.Figure(go.Pie(                                    # assign pie as our first output to return
        name = "",
        values = elements_values,
        labels = elements_labels,
        marker=dict(colors=colors, line=dict(color='#000000', width=2)),# assign the color palette from colors list
        hovertemplate = "%{label}: <br>Amount: %{percent} </br>"
        ))
    elif  (val_pie == "Gender"):
        elements = AC_Villagers["Gender"].value_counts()          # slice data
        elements_labels = elements.index.tolist()                 # get the names for the labels from the index
        elements_values = elements.values                         # to set values from the dataframe
        pie= go.Figure(go.Pie(                                    # assign pie as our first output to return
        name = "", 
        values = elements_values,
        labels = elements_labels,
        marker=dict(colors=colors, line=dict(color='#000000', width=2)),# assign the color palette from colors list
        hovertemplate = "%{label}: <br>Amount: %{percent} </br>"
        ))

    elif  (val_pie == "Style 1"):
        elements = AC_Villagers["Style 1"].value_counts()         # slice data
        elements_labels = elements.index.tolist()                 # get the names for the labels from the index
        elements_values = elements.values                         # to set values from the dataframe
        pie= go.Figure(go.Pie(                                    # assign pie as our first output to return
        name = "",
        values = elements_values,
        labels = elements_labels,
        marker=dict(colors=colors, line=dict(color='#000000', width=2)),# assign the color palette from colors list
        hovertemplate = "%{label}: <br>Amount: %{percent} </br>"
        ))
    elif  (val_pie == "Style 2"):
        elements = AC_Villagers["Style 1"].value_counts()         # slice data
        elements_labels = elements.index.tolist()                 # get the names for the labels from the index
        elements_values = elements.values                         # to set values from the dataframe
        pie= go.Figure(go.Pie(                                    # assign pie as our first output to return
        name = "",
        values = elements_values,
        labels = elements_labels,
        marker=dict(colors=colors, line=dict(color='#000000', width=2)),# assign the color palette from colors list
        hovertemplate = "%{label}: <br>Amount: %{percent} </br>"
        ))
        
        # assign scatterplot as our second output to return
    scatterplot=px.scatter(
    data_frame = AC_Villagers, # data
    x = "Style 1", # x value
    y = "Birthday", # y value
    color = val_variable, # Assign Color Marks 
    facet_row="Gender", # Create a separate row based plot based on this column
    title = "Animal Crossing Villagers"
    ,template=val_themes 
    )
        
    return pie, scatterplot

    
@app.callback(
  Output(component_id='body-div', component_property='children'),
    Input(component_id='show-secret', component_property='n_clicks')
)
def randomize_villager(n_clicks):
    # randomize a villager from 1 to 391
    info_list=["Name","Species","Gender","Personality","Hobby","Birthday",
                                                  "Catchphrase","Favorite Song","Photo Image","House Image","Icon Image"]
    #get their info
    info=list(AC_Villagers.loc[random.randint(1,392),info_list])
    #check if the button was clicked
    if n_clicks is None:
        raise PreventUpdate
    else:                # assign villager info to html table
        return html.Div([
html.Table([
            html.Tr([html.Th('Icon Image:'),
                    html.Td(html.Img(src=info[10]))]),
            html.Tr([html.Th('Name:'),
                     html.Td(info[0])]),
            html.Tr([html.Th('Species:'),
                     html.Td(info[1])]),
            html.Tr([html.Th('Gender:'),
                     html.Td(info[2])]),
            html.Tr([html.Th('Personality:'),
                     html.Td(info[3])]),
            html.Tr([html.Th('Hobby:'),
                     html.Td(info[4])]),
            html.Tr([html.Th('Birthday:'),
                     html.Td(info[5])]),
            html.Tr([html.Th('Catchphrase:'),
                     html.Td(info[6])]),
            html.Tr([html.Th('Favorite Song:'),
                     html.Td(info[7])]),
            html.Tr([html.Th('Photo Image:'),
                     html.Td(html.Img(src=info[8], width=200, height=200))]),
            html.Tr([html.Th('House Image:'),
                     html.Td(html.Img(src=info[9], width=200, height=200))])
        ],style={'table-layout': 'fixed','marginLeft': 'auto', 'marginRight': 'auto','border': 'solid',
  'border-width': '0.2px'})  
 
    ])


@app.callback(
    Output('graph3', 'figure'),
    [Input("bar", "value"),
     Input("top", "value")]
    
)
def update_barplot(song_names, val_top):
    # dictionary for sample songs
    sample_dict={"Sample 1":list(AC_Villagers["Favorite Song"].value_counts().index[0:20]),
               "Sample 2":list(AC_Villagers["Favorite Song"].value_counts().index[21:40]),
               "Sample 3":list(AC_Villagers["Favorite Song"].value_counts().index[41:60])}
    # get which sample was chosen
    for k,v in sample_dict.items():
        if(song_names==k):
            which_sample_list=v
    # dictionary of all songs with their count       
    songs_dict = AC_Villagers["Favorite Song"].value_counts().to_dict()
    # get the count of the selected sample songs
    list_2=[]
    for k,v in songs_dict.items():
        for i in range(0,len(which_sample_list)):
            if(which_sample_list[i]==k):
                list_2.append(v)

    # display all songs checked in radio button
    if(val_top=="All"):
        barplot=px.bar( AC_Villagers,
                 y=list_2,
                 x=which_sample_list,
                 color=which_sample_list, 
                 title="Villagers Favorite Songs",template="ggplot2")
    # display top 3 songs checked in radio button
    elif(val_top=="Top 3"):
        barplot=px.bar( AC_Villagers,
                 y=list_2[0:3],
                 x=which_sample_list[0:3],
                 color=which_sample_list[0:3], 
                 title="Villagers Favorite Songs",template="ggplot2")
# display top 5 songs checked in radio button    
    elif(val_top=="Top 5"):
        barplot=px.bar( AC_Villagers,
                 y=list_2[0:5],
                 x=which_sample_list[0:5],
                 color=which_sample_list[0:5], 
                 title="Villagers Favorite Songs",template="ggplot2")

        
    return barplot


if __name__ == '__main__':
    app.run_server(debug=True) # or whatever you choose
```
